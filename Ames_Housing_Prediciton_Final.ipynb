# Data Cleaning

```python
housing_1_num = housing_1.drop(['Order','PID','Year Remod/Add','Garage Yr Blt'] , axis=1).select_dtypes(include = ['float64', 'int64'])
housing_1_num
```

# Correlation

```python
corr = housing_1_num.corr()['SalePrice']
corr
```

```python
top_features = corr[(corr > 0.1) | (corr < - 0.1)].sort_values(ascending=False) #displays  correlation coefficient greater than 0.5
top_features = top_features.drop('SalePrice')
top_features.index
```

# Features and Target

```python
target = housing_1_num.iloc[: , -1]
target
```

```python
features = housing_1_num.drop('SalePrice' ,axis=1 )
features = features[top_features.index]
features
```

# Splitting the data

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
X_train, X_test, y_train, y_test = train_test_split(features,target,test_size=0.2,random_state=0)

print("Training and testing split was successful.")
```

# Testing base models

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
import xgboost as xgb
from sklearn.svm import SVR
from sklearn.metrics import r2_score

## building different models
def model_building(model_name):
    model = model_name
    model.fit(X_train, y_train)
    print(f"******** Model :- {model_name} ********")
    
    # Predictions for training and test set
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    # Evaluate the model using R² score
    train_r2 = r2_score(y_train, y_train_pred)
    test_r2 = r2_score(y_test, y_test_pred)

    print(f"Training R² Score: {train_r2:.4f}")
    print(f"Testing R² Score: {test_r2:.4f}")
    print()

## Dictionary with different regression models
model_dict = {
    'dt': DecisionTreeRegressor(random_state=0), 
    'knn': KNeighborsRegressor(n_neighbors=17), 
    'rf': RandomForestRegressor(random_state=0), 
    'xgb': xgb.XGBRegressor(random_state=0),
    'svr': SVR()
}

## Calling to build and evaluate models
for key in model_dict.keys():
    model_building(model_dict[key])
```

# BayesianSearch

```python
from skopt import BayesSearchCV
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score , accuracy_score

# Define the hyperparameter search space
param_space = {
    'learning_rate': (0.01, 1.0, 'log-uniform'),
    'max_depth': (2, 10),
    'n_estimators': (100, 500),
    'subsample': (0.5, 1.0),
    'max_features': (0.5, 1.0)
}

# Set up the GradientBoostingRegressor model and Bayesian search
gbr_model = GradientBoostingRegressor(random_state=0)
bayes_search = BayesSearchCV(estimator=gbr_model, search_spaces=param_space, n_iter=30, cv=5, scoring='r2')

# Fit the model to find the best hyperparameters
bayes_search.fit(X_train, y_train)
best_model = bayes_search.best_estimator_

# Predictions for training and testing sets
y_train_pred = best_model.predict(X_train)
y_test_pred = best_model.predict(X_test)

# Calculate R² score (accuracy) for training and testing sets
train_r2 = r2_score(y_train, y_train_pred)
test_r2 = r2_score(y_test, y_test_pred)

# Output the best parameters found
print(bayes_search.best_params_)
print(f"Training R2 Score: {train_r2:.4f}")
print(f"Testing R2 Score: {test_r2:.4f}")
```

# Testing the model on random houses

```python
# Function to check the output for a specific house (regression model)
def check_house_output_by_index(house_index, model, X_test, y_test=None):
    
    # Retrieve the house's data by index
    house_data = X_test.iloc[house_index]

    # Reshape the data to be 2D for prediction
    house_data = house_data.values.reshape(1, -1)

    # Predict the outcome for the house using the best model
    house_prediction = model.predict(house_data)

    # Output the result
    if y_test is not None:
        actual_output = y_test.iloc[house_index]
        difference = abs(house_prediction[0] - actual_output)
        
        # Calculate percentage difference
        percentage_difference = (difference / actual_output) * 100
        
        # Output the result with percentage difference
        print(f"House Index: {house_index}")
        print(f"Prediction: {house_prediction[0]:.2f}, Actual: {actual_output:.2f}, Difference: {difference:.2f}, Percentage Difference: {percentage_difference:.2f}%")
        print('-' * 50)
    
    return house_prediction, actual_output, percentage_difference

# Variables to track houses with percentage differences less than or greater than 5%
less_than_5_percent = 0
less_than_10_percent= 0
less_than_15_percent= 0
greater_than_15_percent = 0

# Loop through test set for a set of house indices
for i in range(0, 440):
    house_index = i   # Adjust the index increment as needed
    house_prediction, actual_output, percentage_difference = check_house_output_by_index(house_index, best_model, X_test, y_test)

    # Conditional to count houses with percentage differences
    if percentage_difference <= 5:
        less_than_5_percent += 1
    elif percentage_difference <= 10:
        less_than_10_percent += 1
    elif percentage_difference <= 15:
        less_than_15_percent += 1
    else:
        greater_than_15_percent += 1

# Display the count of houses with percentage differences
print()
print(f"Number of houses with prediction difference <= 5%: {less_than_5_percent}")
print(f"Number of houses with 5% < prediction difference <= 10%: {less_than_10_percent}")
print(f"Number of houses with 10% < prediction difference <= 15%: {less_than_15_percent}")
print(f"Number of houses with prediction difference > 15%: {greater_than_15_percent}")
```
